#### Now we isolate the signal to eccentric and concentric movements. (Up and down motion)
We do this taking into account the marker position, as it rises and falls, we find certain peaks, which correspond to the end of the rising motion and the valleys correspond to the end of the falling motion. We then figure out what time does this correlate to, and we plot each of the phases of the press-up cycle.
- Eccentric Phase lowering the body towards the ground, muscles lengthening under tension.
- Concentric Phase pushing the body back up, muscles contracting and shortening to generate force.

We can investigate the relationship between muscle activation and force exertion through calculating the correlation between the sEMG signals and the force plate data.

```python
# Normalize sEMG data
semg_data_normalized_left = df[['Left Ser Ant', 'Left Pec Maj', 'Left Delt Ant', 'Left Triceps']] / df[['Left Ser Ant', 'Left Pec Maj', 'Left Delt Ant', 'Left Triceps']].max()

semg_data_normalized_right = df[['Right Ser Ant', 'Right Pec Maj', 'Right Delt Ant', 'Right Triceps']] / df[['Right Ser Ant', 'Right Pec Maj', 'Right Delt Ant', 'Right Triceps']].max()

# Normalize force_z data
force_data_normalized_left = df['Fz Left'] / df['Fz Left'].max()

force_data_normalized_right = df['Fz Right'] / df['Fz Right'].max()
```

```python
# Calculate the correlation for the left side
correlation_left = semg_data_normalized_left.corrwith(force_data_normalized_left)

# Calculate the correlation for the right side
correlation_right = semg_data_normalized_right.corrwith(force_data_normalized_right)
```

```python
# Create a new DataFrame containing both correlations
correlation_df = pd.concat([correlation_left.rename('Left'), correlation_right.rename('Right')], axis=1)

# Plot the correlation for both sides
plt.figure()
correlation_df.plot(kind='bar', alpha=0.5)
plt.title('Correlation between sEMG Signals and Forces')
plt.xlabel('Muscle')
plt.ylabel('Correlation Coefficient')
plt.legend(loc='best')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('BiomechRehab2023Pt2.csv.csv')

print(df)
```

```python
# Define the RMS filter function
def rms_filtering(sEMG_FP_signal, window_size=200, resampling_rate=120):
    squared_signal = np.square(sEMG_FP_signal)
    windowing = np.ones(window_size) / window_size
    filtered_signal = np.convolve(squared_signal, windowing, mode='valid')
    rms_sEMG_FP_signal = np.sqrt(filtered_signal)

    # Resample the signal at the desired rate
    resampled_signal = np.interp(
        np.linspace(0, len(rms_sEMG_FP_signal) - 1, int(len(rms_sEMG_FP_signal) * resampling_rate / 1000)),
        np.arange(len(rms_sEMG_FP_signal)),
        rms_sEMG_FP_signal,
    )

    return resampled_signal

# Apply the RMS filter to each column of the dataset
filtered_data = df.apply(lambda column: rms_filtering(column, window_size = 250))
```

```python
import numpy as np
import scipy.signal

def MNF(sEMG_signal, fs = 120, k=15, overlap=0.5):
    nperseg = len(sEMG_signal) // k
    noverlap = int(nperseg * overlap)
    
    # Applying Hamming window as outlined in paper
    window = scipy.signal.windows.hamming(nperseg)
    
    # Estimate power spectral density (PSD) using Welch method, fs value needs to be input
    f, Pxx = scipy.signal.welch(sEMG_signal, fs, window=window, nperseg=nperseg, noverlap=noverlap)
    
    # Calculate mean frequency (MNF) 
    mnf_numerator = np.sum(f * Pxx)
    mnf_denominator = np.sum(Pxx)
    mnf = mnf_numerator / mnf_denominator
    
    return mnf

# In order to not just only calculate a single value we need to slide the filter, in order to see the change over time.
def MNF_sliding_window(sEMG_signal, window_size=200, step_size=100, fs=120, k=15, overlap=0.5):
    mnfs = []
    for i in range(0, len(sEMG_signal)-window_size, step_size):
        sEMG_window = sEMG_signal[i:i+window_size]
        mnf = MNF(sEMG_window, fs=fs, k=k, overlap=overlap)
        mnfs.append(mnf)
    return mnfs

# Example usage
fs = 120
time = filtered_data['Time (secs)']
lp = filtered_data['Left Pec Maj']
rp = filtered_data['Right Pec Maj']

ld = filtered_data['Left Delt Ant']
rd = filtered_data['Right Delt Ant']

ls = filtered_data['Left Ser Ant']
rs = filtered_data['Right Ser Ant']

lt = filtered_data['Left Triceps']
rt = filtered_data['Right Triceps']

lf = filtered_data['Fz Left']
rf = filtered_data['Fz Right']

mean_frequency_lp = MNF_sliding_window(lp)
mean_frequency_rp = MNF_sliding_window(rp)

mean_frequency_ld = MNF_sliding_window(ld)
mean_frequency_rd = MNF_sliding_window(rd)

mean_frequency_ls = MNF_sliding_window(ls)
mean_frequency_rs = MNF_sliding_window(rs)

mean_frequency_lt = MNF_sliding_window(lt)
mean_frequency_rt = MNF_sliding_window(rt)

mean_frequency_lf = MNF_sliding_window(lf)
mean_frequency_rf = MNF_sliding_window(rf)

# Plot the MNF values over time
t_mnfs = np.arange(0, len(mean_frequency_lp)*100, 100)
t = [0, 10, 20 ,30]
t_current = [0,2000,4000,6000]

plt.xticks(t_current, t)
plt.plot(t_mnfs, mean_frequency_lp)
plt.xlabel('Time (s)')
plt.ylabel('MNF')
plt.show()
```
